## I. Introduction
This project is a simple, beginner-friendly Content Management System (CMS) designed to create and manage **website pages** and **blog posts**, similar to a small-scale WordPress or Drupal.  
It provides a public website and an internal dashboard to manage content, navigation menus, and basic website settings.

---

## II. Objectives of the Project
The CMS is designed to help end users (non-technical) to:  
- Easily create and publish new pages on their website (e.g., About, Contact).  
- Write and manage blog posts without needing to code.  
- Organize pages into a navigation menu visible to visitors.  
- Save drafts and control which content is published or unpublished.  
- Maintain a professional-looking website with minimal effort.

---

## III. Target Audience
The CMS is intended for:  
- Small business owners who want to manage their website content themselves.  
- Bloggers or content creators who want to post updates without technical knowledge.  
- Organizations or personal websites needing simple page management.  
- Users who need a beginner-friendly alternative to complex CMS platforms.

---

## IV. Key Features

### 1. Dashboard
- Overview of website content and recent activity.  
- Quick access to create new pages or blog posts.

### 2. Page Management
- Create, edit, publish, or unpublish website pages.  
- Optional display in the main navigation menu.  
- Delete pages (Admin only).

### 3. Blog Post Management
- Create and edit blog posts.  
- Save drafts and publish content.  
- Optional categories/tags for organization.  
- Delete posts (Admin only).

### 4. Menu Management (Admin Only)
- Add pages to the website navigation.  
- Reorder menu items.  
- Hide pages from the menu without unpublishing.

### 5. Basic Website Settings (Admin Only)
- Set site title and tagline.  
- Upload logo.  
- Choose default homepage (static page or blog list).

### 6. User Roles
- **Admin:** full access to all features.  
- **Editor:** can manage pages and posts but cannot manage users or settings.  
- **Public Visitor:** can view only published content.

### 7. Public Website
- Clean homepage with site logo and menu.  
- Accessible static pages.  
- Blog section with published posts.  
- Individual pages for each static page or blog post.

## VI. System Architecture

### 1. Public Website
- Shows published pages and blog posts.  
- Accessible to all visitors.  
- Navigation menu reflects dashboard configuration.

### 2. Admin Dashboard
- Accessible only to logged-in users.  
- Admins have full control; Editors have limited access.  
- Sections include Pages, Blog Posts, Menu, and Settings.

### 3. Content Workflow
- Draft → Published → Visible on public site.  
- Editors can edit content; Admins can publish, unpublish, or delete content.  

### 4. Role-Based Access
- Admins manage everything.  
- Editors focus on content creation.  
- Public users only read published content.
